```python
import boto3
import pandas as pa
import cv2
import matplotlib.pyplot as plt
```

```python
# load keys from vault
AWS_KEY_ID = pa.read_csv('../../AWS_vault/credentials_boto_test_2.csv')['Access key ID'][0]
AWS_SECRET = pa.read_csv('../../AWS_vault/credentials_boto_test_2.csv')['Secret access key'][0]

# define your preferred region
my_region = 'eu-central-1'
```

```python
rekog = boto3.client('rekognition',  
                     region_name= my_region,   
                     aws_access_key_id=AWS_KEY_ID,   
                     aws_secret_access_key=AWS_SECRET)

s3 = boto3.client('s3',  region_name= my_region,
                         aws_access_key_id=AWS_KEY_ID, 
                         aws_secret_access_key=AWS_SECRET)
```

```python
image_1 = cv2.imread('./images/21NYTODAY01-superJumbo.jpg')
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
plt.imshow(image_1)
```

```python
s3.upload_file(Filename='images/21NYTODAY01-superJumbo.jpg',
               Bucket='test-bokeh', 
               Key='image/image')

response = rekog.detect_labels(Image={'S3Object': 
                                              {'Bucket': 'test-bokeh', 
                                               'Name': 'image/image'}}, 
                               MaxLabels=1)
```

```python
print(response['Labels'])
```

```python
car_count = 0
boxes=[]
confidence_level = 80

# Iterate over the labels in the response
for label in response['Labels']:
    # Find the cat label, look over the detected instances
    if label['Name'] == 'Car':
        for instance in label['Instances']:
            # Only count instances with confidence_level
            if (instance['Confidence'] > confidence_level):
                car_count += 1
                boxes.append(instance['BoundingBox'])
                
print('Total number of cars found: {}'.format(car_count), '\n', boxes)
```

```python
# get bounding boxes and plot them 
image_1 = cv2.imread('images/21NYTODAY01-superJumbo.jpg')
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
im_h,im_w,_ = image_1.shape

for i in boxes:
    y=int((im_h * i['Top']))
    x=int((im_w * i['Left']))
    h=int((im_h * i['Height']))
    w=int((im_w * i['Width']))
    # add bounding boxes to image
    cv2.rectangle(image_1,(x,y),(x+w,y+h),(255,0,0),10)

    plt.imshow(image_1)
```

```python
plt.imsave('21NYTODAY01-superJumbo_with_boxes.png', image_1)
```

```python
# get bounding boxes and plot them 
image_1 = cv2.imread(filenames[4])
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
im_h,im_w,_ = image_1.shape

for i in boxes:
    y=int((im_h * i['Top']))
    x=int((im_w * i['Left']))
    h=int((im_h * i['Height']))
    w=int((im_w * i['Width']))
    # add bounding boxes to image
    cv2.rectangle(image_1,(x,y),(x+w,y+h),(255,0,0),10)

    plt.imshow(image_1)
```

```python

def overlap(image_filename, box_cars, box_bike): 
    image_1 = cv2.imread(image_filename)
    im_h,im_w,_ = image_1.shape

    counter=0
    over = {}
    for b in box_bike:
        l1y=int((im_h * b['Top']))
        l1x=int((im_w * b['Left']))
        l1h=int((im_h * b['Height']))
        l1w=int((im_w * b['Width']))
        for c in box_cars:   
            l2y=int((im_h * c['Top']))
            l2x=int((im_w * c['Left']))
            l2h=int((im_h * c['Height']))
            l2w=int((im_w * c['Width']))

            # bike on the left side of the car
            if l1x < l2x:
                # check if boxes overlap
                if(l1x + l1w > l2x and l1y+l1h > l2y):
                    print("bike on the car's felt size.")
                    counter+=1
                    over.update({counter:[[l1x,l1y,l1h,l1w],[l2x,l2y,l2h,l2w]]})
                else:
                    print('No overlapping cars and bikes')
            # bike on the right side of the car 
            elif l1x > l2x: 
                if(l2x + l2w > l1x and l2y+l2h > l1y):
                    print("bike of the car's right side.")
                    counter+=1
                    over.update({counter:[[l1x,l1y,l1h,l1w],[l2x,l2y,l2h,l2w]]})
                else:
                    print('No overlapping cars and bikes')
            else: 
                print('No overlapping cars and bikes')
    return over

    
def check_if_car_or_bike_missing(box_cars, box_bike):
    if box_cars==[]:
        print('No cars')
    elif box_bike==[]:
        print('No bikes')
    else:
        pass #overlap()

def recognit(image_filename):
    s3.upload_file(Filename=image_filename,
               Bucket='test-bokeh', 
               Key=image_filename)

    response = rekog.detect_labels(Image={'S3Object': 
                                           {'Bucket': 'test-bokeh', 
                                              'Name': image_filename}}, 
                                    MaxLabels=123, MinConfidence=90)
    return response    
    

def obj_classifier(image_filename):
    box_cars = []
    box_bike = []
    confidence_level = 90
    response = recognit(image_filename)
                        
    # Iterate over the labels in the response
    for label in response['Labels']:
        for instance in label['Instances']:
            if label['Name']=='Bicycle':
                box_bike.append(instance['BoundingBox'])
            elif label['Name']=='Car':
                box_cars.append(instance['BoundingBox'])       
    # returns dict with classified boxes 
    return {'box_bike':box_bike, 'box_cars':box_cars}
               
def overlapping_boxes(image_filename):
    oc = obj_classifier(image_filename)    
    overlapping = overlap(image_filename, oc['box_cars'], oc['box_bike'])
    #print("Number of overlapping cars and bikes: {}".format(n_overlapping))
    return overlapping


def plot_overlap(overlapping, image_filename):
    image_1 = cv2.imread(image_filename)
    image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
    im_h,im_w,_ = image_1.shape
    for item in overlapping.items():
        l1x, l1y, l1h, l1w = item[1][0]
        cv2.rectangle(image_1,(l1x,l1y),(l1x+l1w, l1y+l1h),(255,0,0),10)
        l2x, l2y, l2h, l2w = item[1][1]
        cv2.rectangle(image_1,(l2x,l2y),(l2x+l2w, l2y+l2h),(255,0,0),10)
    plt.imshow(image_1)
    
    
# furhter ideas
def accident_report(image_filename):
    pass
    
def crossroad_traffic(image_filename):
    pass

def parking_place(image_filename):
    pass
    
def cars_near_far():
    pass
            
```
